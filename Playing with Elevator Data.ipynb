```python
import pandas as pd
import pymysql
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_list(queue):
    result = []
    tmp = []
    
    while(len(queue)>0):
        e = queue.pop(0)
        if ~np.isnan(e):
            tmp.append(e)
            # print(e)
        elif len(tmp)>0:
            result.append(tmp)
            tmp = []
            # print(result)
        else:
            continue
   
    return result

def get_index(queue):
    result = []
    tmp = []
    # print(len(queue))
    
    for i in range(len(queue)):
        e = queue.pop(0)
        if ~np.isnan(e):
            tmp.append(i)
            # print(i)
        elif len(tmp)>0:
            result.append(tmp)
            tmp = []
            # print(result)
        else:
            continue
   
    return result
```

```python
Item005 = list(df['Item005'])
result = get_list(Item005)
print(result)
r = np.array(result)
print(r.shape)
for i in range(len(r)):
    print(len(r[i]))
```

```python
Item005 = list(df['Item005'])
result_i = get_index(Item005)
# print(result_i)
ln = []
for i in range(len(result_i)):
    a = result_i[i]
    if len(a) < 50:
        continue
    else:
        print("Step: ", i, "len: ", len(a))
        # print(a)
        print("Start Index:", a[0], "Stop Index:", a[-1])
        stti = a[0]
        stpi = a[-1]
        tmp_df = df.iloc[stti:stpi]
        print(tmp_df.shape)
        print(tmp_df)
        start = tmp_df['DataSavedTime'].iloc[0]
        stop = tmp_df['DataSavedTime'].iloc[-1]
        print("Start:", tmp_df['DataSavedTime'].iloc[0])
        print("Stop:", tmp_df['DataSavedTime'].iloc[-1])
        print(stop - start)
        plt.plot(tmp_df['DataSavedTime'], tmp_df['Item005'])
        plt.show()
        ln.append(len(a))
    
plt.hist(ln)
print(len(ln))
```
