```python
import pandas as pd
import numpy as np
import pymysql
import matplotlib.pyplot as plt
import os
from pandas.tseries import converter
converter.register()
```

```python
conn = pymysql.connect(
    host='192.168.100.225',
    port=3306,
    user='root',
    password='its@1234',
    db='UYeG_KUMHO',
    charset='utf8'
)

curs = conn.cursor()

set_day1 = '2019-10-11 00:00:00'
set_day2 = '2019-10-12 23:59:59'


# LAST 24HOURS : 2019-10-09 10:00:00 ~ 2019-10-10 10:00:00 : 10초 데이터 가져오기
sql_current = "SELECT DataSavedTime, Item001, Item002, Item003, Item004, Item005, \
    Item006, Item007, Item008 FROM HisItemCurrSecond WHERE DataSavedTime \
    BETWEEN '"+set_day1+"' and '"+set_day2+"' ORDER BY DataSavedTime ASC"

sql_waterlevel = "SELECT DataSavedTime, Item011, Item012 \
    FROM HisItemWaterLevelSecond WHERE DataSavedTime \
    BETWEEN '"+set_day1+"' and '"+set_day2+"' ORDER BY DataSavedTime ASC"

sql_feed = "SELECT DataSavedTime, Item009, Item010 \
    FROM HisItemFreqSecond WHERE DataSavedTime \
    BETWEEN '"+set_day1+"' and '"+set_day2+"' ORDER BY DataSavedTime ASC"

curs.execute(sql_current)
rows_current=curs.fetchall()

curs.execute(sql_waterlevel)
rows_waterlevel=curs.fetchall()

curs.execute(sql_feed)
rows_feed=curs.fetchall()

conn.close()

data_current = list(rows_current)
df_current = pd.DataFrame(data_current)
df_current.columns = ['DataSavedTime', 'Item001', 'Item002', 'Item003', 'Item004', 'Item005', 'Item006', 'Item007', 'Item008']

data_waterlevel = list(rows_waterlevel)
df_waterlevel = pd.DataFrame(data_waterlevel)
df_waterlevel.columns = ['DataSavedTime', 'Item011', 'Item012']

data_feed = list(rows_feed)
df_feed = pd.DataFrame(data_feed)
df_feed.columns = ['DataSavedTime', 'Item009', 'Item010']


print(df_current)
print(df_waterlevel)
print(df_feed)
```

```python
df = df_current.join(df_feed.set_index('DataSavedTime'), on='DataSavedTime')
# df.columns = ['DataSavedTime', 'Item001', 'Item002', 'Item003', 'Item004', 'Item005', 'Item006', 
#              'Item007', 'Item008' , 'Item009', 'Item010']

df = df.join(df_waterlevel.set_index('DataSavedTime'), on='DataSavedTime')
# df.columns = ['DataSavedTime', 'Item001', 'Item002', 'Item003', 'Item004', 'Item005', 'Item006', 
#              'Item007', 'Item008' , 'Item009', 'Item010' , 'Item011', 'Item012']
print(df.head())
```

```python
motor_name = ["Wash_Pump01", "Wash_Pump02", "Rinse_Pump05", "Wash_Dryer01", 
              "Rinse_Pump04", "Wash_Pump03", "Rinse_Dryer02", "Rinse_Dryer03", "Feed_01", "Feed_02", "Tank_#21", "Tank_#22"]
colors = ["red", "orange", "yellow", "green", "blue", "navy", "purple", "cyan", "black", "magenta", "gray", "violet"]

def paint(date, data, motor_name, color):
	plt.style.use('ggplot')
	plt.figure(figsize=(15,5))
	# plt.ylim(0,40)
	plt.title(motor_name)
	plt.plot(date, data, color, label=motor_name)
	plt.grid(True)
	plt.xlabel('Time')
	plt.ylabel('Value (I)')
	plt.legend()
	output_path = os.path.abspath("plots")
	# plt.savefig(os.path.join(output_path, str(date[0])[:10]+"_"+motor_name+".png"))
	plt.show()
    
paint(df["DataSavedTime"], df["Item001"], motor_name[0], colors[0])
```
