```python
import pandas as pd
import pymysql
import numpy as np
import matplotlib.pyplot as plt
```

defs

```python
def get_list(queue):
    result = []
    tmp = []
    
    while(len(queue)>0):
        e = queue.pop(0)
        if ~np.isnan(e):
            tmp.append(e)
            # print(e)
        elif len(tmp)>0:
            result.append(tmp)
            tmp = []
            # print(result)
        else:
            continue
   
    return result

def get_index(queue):
    result = []
    tmp = []
    # print(len(queue))
    
    for i in range(len(queue)):
        e = queue.pop(0)
        if ~np.isnan(e):
            tmp.append(i)
            # print(i)
        elif len(tmp)>0:
            result.append(tmp)
            tmp = []
            # print(result)
        else:
            continue
   
    return result
```
